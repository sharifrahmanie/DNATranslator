```python
def dana_translate_3frame(DNA):
    
    '''
    This is a simple DNA trasnlator that use a dictionary for traslating forward strand of DNA.
    @Author: https://github.com/sharifrahmanie
    '''
    
    # Making the noncoding starand, which is going to transcrid to mRNA
    noncoding_strand=[]
    for t in DNA:
        if t == "A":
            noncoding_strand.append('T')
        elif t == "C":
            noncoding_strand.append('G')
        elif t == "G":
            noncoding_strand.append('C')
        elif t == "T":
            noncoding_strand.append('A') 
    
    # putting together the noncoding starand, because they are a list of comma separated ['A', 'C',..]
    Noncoding_Strand=''.join(noncoding_strand)
    
    # Translation to mRNA
    mrna=[]
    for t in Noncoding_Strand:
        if t == "A":
            mrna.append('U')
        elif t == "C":
            mrna.append('G')
        elif t == "G":
            mrna.append('C')
        elif t == "T":
            mrna.append('A') 
    
    # putting together the mRNA seq
    mRNA=''.join(mrna)
     
        
        
        
    # Collecting nuclotides as list
    code=[]
    
    # Collecting codons for first frame
    code1=[]
    
    # Collecting codons for second frame
    code2 = []
    
    # Collecting codons for third frame
    code3 = []
    
    # Collecting AA
    protein_first_frame = []
    
    
    # Collecting AA
    protein_second_frame = []
    
    # Collecting AA
    protein_third_frame = []
    
    # Making a dictionery of codons and their associated AA
    codons = {
            'AUG':'M',
            'UUU':'F', 'UUC':'F', 'UUA':'L', 'UUG':'L', 'CUU':'L', 'CUC':'L', 'CUA':'L', 'CUG':'L' ,
            'AUU':'I', 'AUC':'I', 'AUA':'I', 'GUU':'V', 'GUC':'V', 'GUA':'V', 'GUG':'V', 'UAU':'Y',
            'UAC':'Y', 'UAA':'STOP', 'UAG':'STOP', 'UGA':'STOP', 
            'CAU':'H', 'CAC':'H', 'CAA':'Q', 'CAG':'Q', 'AAU':'N', 'AAC':'N', 'AAA':'K', 'AAG':'K', 
            'GAU':'D', 'GAC':'D', 'GAA':'E', 'GAG':'E', 'UGU':'C', 'UGC':'C', 'UGG':'W', 
            'CGU':'R', 'CGC':'R', 'CGA':'R', 'CGG':'R', 'AGA':'R', 'AGG':'R', 'UCU':'S', 'UCC':'S', 
            'UCA':'S', 'UCG':'S', 'AGU':'S', 'AGC':'S', 
            'GGU':'G', 'GGC':'G', 'GGA':'G', 'GGG':'G', 'CCU':'P', 'CCC':'P', 'CCA':'P', 'CCG':'P',
            'GCU':'A', 'GCC':'A', 'GCA':'A', 'GCG':'A', 'ACU':'T', 'ACC':'T', 'ACA':'T', 'ACG':'T'
        } 
        
   
    # Iteration over mRNA seq to make nuclotides as a list of comma separated for further
    #  proccessing ['A','C','G',...]
    for i in mRNA:
        for j in i:
            code.append(j)
              
    
    # Preparing the mRNA as separated codons ['ACG','CUC',...]    
    while len(code) >= 3:
        # Appending each three nuclotides at each iteration
        code1.append(''.join(code[0:3]))
        # After adding each codon we need to remove its nuclotides
        code.pop(0)
        code.pop(0)
        code.pop(0)
        
        
        
    # Make our list empty for another frame
    code.clear()
        
    # For second frame we need to fill our code list again   
    for i in mRNA:
        for j in i:
            code.append(j)   
        
    # Removing the frist nuclotide to strat at the second frame
    code.pop(0)
    while len(code) >= 3:
        # Appending each three nuclotides at each iteration
        code2.append(''.join(code[0:3]))
        # After adding each codon we need to remove its nuclotides
        
        code.pop(0)
        code.pop(0)
        code.pop(0)    

    # Make our list empty for another frame
    code.clear()    
        
    # For third frame we need to fill our code list again   
    for i in mRNA:
        for j in i:
            code.append(j)   
            
    # Removing the two frist nuclotides to strat at the third frame        
    code.pop(0)
    code.pop(0) 
    
    # Making codons ready for third frame traslation  
    while len(code) >= 3:
        
        # Appending each three nuclotides at each iteration
        code3.append(''.join(code[0:3]))
        code.pop(0)
        code.pop(0)
        code.pop(0) 
    
    
    # Finding AA for first frame
    for c in code1:
        for b in codons:
            if b==c:
                protein_first_frame.append(codons[c])
            elif c in ['UAA', 'UAG', 'UGA']:
                protein_first_frame.append("*")
                break
    protein_FF =''.join(protein_first_frame)
    
    
      # Finding AA for second frame
    for c in code2:
        for b in codons:
            if b==c:
                protein_second_frame.append(codons[c])
            elif c in ['UAA', 'UAG', 'UGA']:
                protein_second_frame.append("*")
                break
    protein_SF =''.join(protein_second_frame)
    
    
      # Finding AA for third frame
    for c in code3:
        for b in codons:
            if b==c:
                protein_third_frame.append(codons[c])
            elif c in ['UAA', 'UAG', 'UGA']:
                protein_third_frame.append("*")
                break
    protein_TF =''.join(protein_third_frame)
    
    
    
    
    
    
    
    print("Translation in the fist frame (forward): ","\t", protein_FF)
    print("Translation in the second frame (forward): ","\t", protein_SF)
    print("Translation in the third frame (forward): ","\t", protein_TF) 
```
